### Customer Life Time Value of an Automobile Company

```python
import pandas as pd
import numpy as np
pd.set_option('display.max_columns', 500)
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import seaborn as sns
```

```python
df = pd.read_csv('WA_Fn-UseC_-Marketing-Customer-Value-Analysis.csv')
```

```python
df.head()
```

### Exploring data

```python
df.describe()
```

```python
df['Customer Lifetime Value'].hist(figsize=(8,5), grid=False, xlabelsize =12, ylabelsize = 12)
plt.title('Customer Lifetime Value distribution')
plt.show()
```

```python
sns.pairplot(df)
```

```python
df.head()
```

```python
df["Coverage"].unique()
```

### It appears that Customer Life Time Value depends more on Monthly Premium Auto with a positive pearson's correlation coefficient of 0.4

```python
sns.jointplot("Monthly Premium Auto", "Customer Lifetime Value",data=df,  kind='scatter', color="red")
```

```python
sns.jointplot("Total Claim Amount", "Customer Lifetime Value",data=df,  kind='scatter', color="blue")
```

### Checking some categorical variables

```python
sns.barplot(x="Marital Status", y="Customer Lifetime Value", data=df)
```

```python
sns.barplot(x="Gender", y="Customer Lifetime Value", data=df)
```

```python
sns.barplot(x="Location Code", y="Customer Lifetime Value", data=df)
```
